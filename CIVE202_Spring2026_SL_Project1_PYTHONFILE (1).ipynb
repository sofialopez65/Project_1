## Question #1: What are the 5 locations in Nebraska with the highest mean and median concentrations of VOC, PM 2.5 and PM 10.0?

```python
summary = df.groupby("sensor.name").agg({
    "voc": ["mean", "median"],
    "pm2.5_atm": ["mean", "median"],
    "pm10.0_atm": ["mean", "median"]})
# VOC
top5_voc_mean = summary.sort_values(("voc", "mean"), ascending=False).head(5)
top5_voc_median = summary.sort_values(("voc", "median"), ascending=False).head(5)
# PM2.5
top5_pm25_mean = summary.sort_values(("pm2.5_atm", "mean"), ascending=False).head(5)
top5_pm25_median = summary.sort_values(("pm2.5_atm", "median"), ascending=False).head(5)
# PM10.0
top5_pm10_mean = summary.sort_values(("pm10.0_atm", "mean"), ascending=False).head(5)
top5_pm10_median = summary.sort_values(("pm10.0_atm", "median"), ascending=False).head(5)

display(top5_voc_mean[[("voc", "mean")]])
display(top5_voc_median[[("voc", "median")]])
display(top5_pm25_mean[[("pm2.5_atm", "mean")]])
display(top5_pm25_median[[("pm2.5_atm", "median")]])
display(top5_pm10_mean[[("pm10.0_atm", "mean")]])
display(top5_pm10_median[[("pm10.0_atm", "median")]])
                                        
```

## Question #3: Does humidity and temperature have a noticeable effect on air quality?

```python
def humidity_category(h):
    if h < 50:
        return "Low"
    elif h <= 80:
        return "High"
    else:
        return "Very High"

df["humidity_cat"] = df["humidity"].apply(humidity_category)
df["temp_cat"] = df["temperature"].apply(temp_category)

def temp_category(t):
    if t < 32:
        return "Below freezing"
    elif t <= 50:
        return "Cool"
    elif t <= 70:
        return "Warm"
    else:
        return "Hot"

df["temp_cat"] = df["temperature"].apply(temp_category)
summary = df.groupby(["sensor.name", "humidity_cat", "temp_cat"])[["voc","pm2.5_atm","pm10.0_atm"]].mean()
display(summary)
```

## Question #4: Have there been any Air Quality Index (AQI) health risks (unhealthy for sensitive populations) at any of the locations in the dataset for PM 2.5 and PM 10 based on the EPA’s AQI ratings?

```python
df["pm25_risk"] = df["pm2.5_atm"] >= 35.5
df["pm10_risk"] = df["pm10.0_atm"] >= 155
dates_risk = df[(df["pm25_risk"]) | (df["pm10_risk"])]
risk_summary = dates_risk.groupby("sensor.name")[["pm2.5_atm", "pm10.0_atm"]].max()

display(risk_summary)
display(dates_risk[["date", "sensor.name", "pm2.5_atm", "pm10.0_atm"]])
```
